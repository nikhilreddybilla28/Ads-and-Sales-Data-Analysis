```
 import numpy as np
 import pandas as pd
```

```
df = pd.read_excel('MSdata.xlsx','Assignment-2',index_col=0)
df.head()
```

```
df=df.transpose()
print(df.shape)
df.head()
```

```
df['year'] = pd.to_datetime(df.index).year
df['monthly'] = pd.to_datetime(df.index).month
```

```
df
```

```
monthly19 = df[df.year==2019].groupby(['monthly']).sum()
monthly19
```

```
monthly20 = df[df.year==2020].groupby(['monthly']).sum()
monthly20
```

**As the months are not repeated we can neglect year wise separation**

```
df=df.drop(['year'], axis = 1) 
df.head()
```

```
monthly= df.groupby(['monthly']).sum()
monthly
```

```
monthly['total'] = monthly.sum(axis=1)
monthly
```

## **1) What is the monthly & quaterly sales data by SKU**

```
monthly['total']
```

```
print('quarterly:')
print('Q1(jan-mar) ---',monthly['total'][1]+monthly['total'][2]+monthly['total'][3])
print('Q2(apr-jun) ---',monthly['total'][4]+monthly['total'][5]+monthly['total'][6])
print('Q3(jul-sept) ---',monthly['total'][7]+monthly['total'][8]+monthly['total'][9])
print('Q4(oct-dec) ---',monthly['total'][10]+monthly['total'][11]+monthly['total'][12])
```
